```python
# Unificando os arquivos
emprestimos = pd.concat(  [emprestimos_20101, 
            emprestimos_20102 ,
            emprestimos_20111 ,
            emprestimos_20112,
            emprestimos_20121,
            emprestimos_20122,
            emprestimos_20131,
            emprestimos_20132,
            emprestimos_20141,
            emprestimos_20142,
            emprestimos_20151,
            emprestimos_20152,
            emprestimos_20161,
            emprestimos_20162,
            emprestimos_20171,
            emprestimos_20172,
            emprestimos_20181,
            emprestimos_20182,
            emprestimos_20191,
            emprestimos_20192,
            emprestimos_20201],

            axis=0,
            verify_integrity=False,
            ignore_index=False)
```

```python
emprestimos
```

Removendo os dados duplicados

```python
if emprestimos. duplicated().sum()>=0:
    print('Temos dados duplicados')
    emprestimos_sem_duplicados=emprestimos.drop_duplicates()
    quantidade_duplicados = len(emprestimos) - len(emprestimos_sem_duplicados)
    print(f'Foram removidos {quantidade_duplicados} dados duplicados')

else:
    print('Não temos dados duplicados')
    emprestimos_sem_duplicados=emprestimos
```
