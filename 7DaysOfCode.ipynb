```python
import pandas as pd

# Função para categorizar a 'localizacao'
def categorizar_localizacao(valor):
    if 0 <= valor < 100:
        return 'Generalidades. Ciência e conhecimento.'
    elif 100 <= valor < 200:
        return 'Filosofia e psicologia'
    elif 200 <= valor < 300:
        return 'Religião'
    elif 300 <= valor < 400:
        return 'Ciências sociais'
    elif 400 <= valor < 500:
        return 'Classe vaga. Provisoriamente não ocupada'
    elif 500 <= valor < 600:
        return 'Matemática e ciências naturais'
    elif 600 <= valor < 700:
        return 'Ciências aplicadas'
    elif 700 <= valor < 800:
        return 'Belas artes'
    elif 800 <= valor < 900:
        return 'Linguagem. Língua. Linguística'
    elif 900 <= valor < 1000:
        return 'Geografia. Biografia. História'
    else:
        return None

# Aplicar a função a cada valor na coluna 'localizacao'
emprestimos_completo['classe geral na CDU'] = emprestimos_completo['localizacao'].apply(categorizar_localizacao)
```
